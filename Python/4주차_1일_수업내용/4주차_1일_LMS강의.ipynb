# 값의 연결

```python
a = pd.DataFrame(np.arange(1, 10).reshape(3,3))
a
```

```python
b =pd.Series(np.arange(10, 40, 10))
b
```

```python
pd.concat([a,b], axis=1)  # a, b 를 붙힐건데 축은 1(옆으로 하겠다.) concat
```

```python
pd.concat([a,b], axis=1, ignore_index=True)  # 위 컬럼보면 붙힌게 0으로 나와 보기 싫다 자연스럽게 3으로 ignore_index=True
```

```python
a.append(b, ignore_index=True)  # a 에 b를 추가 할건데 밑으로 append
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y_ = np.cos(x)

plt.plot(x, y, label='sin')
plt.plot(x, y_, '-o', label='cos')

plt.legend(loc=4) # loc=4 범례 위치를 숫자로 바꾼다.
```
