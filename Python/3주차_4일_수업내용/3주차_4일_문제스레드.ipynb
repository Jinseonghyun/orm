문제 2. 다음에 정보가 담겨져 있는 것을 Class로 만들어주세요. 메서드로는 '할인' 메서드를 만들어주세요. 예를 들어 아래 책을 10% 할인한다고 하면 4500원이 됩니다. 책.할인(10) == 4500의 형태로 만들어주시면 됩니다.

```python
# 문제 2
 {
    '책제목': 'hello world python',
    '책 내용 요약': 'hello world...',
    '저자': '이호준',
    '발행일': '2024-01-11',
    '가격': 5000
}
```

```python
# 문제 2번 정답
from datetime import datetime

class Book:
    def __init__(self, title, summary, author, publish_date, price):
        self.title = title
        self.summary = summary
        self.author = author
        self.publish_date = publish_date
        self.price = price

    def 할인(self, discount_percentage):
        discounted_price = int(self.price * (1 - discount_percentage / 100))
        return discounted_price

# 주어진 정보로 Book 인스턴스 생성
book_info = {
    '책제목': 'hello world python',
    '책 내용 요약': 'hello world...',
    '저자': '이호준',
    '발행일': '2024-01-11',
    '가격': 5000
}

book_instance = Book(
    title=book_info['책제목'],
    summary=book_info['책 내용 요약'],
    author=book_info['저자'],
    publish_date=datetime.strptime(book_info['발행일'], '%Y-%m-%d').date(),
    price=book_info['가격']
)

# 할인 메서드 테스트
discounted_price = book_instance.할인(10)
print(f"{book_instance.title}을(를) 10% 할인하면 {discounted_price}원입니다.")
```

문제 *3*.   [게임 밸런스 조절](https://100.pyalgo.co.kr/?page=10)
- 게임 밸런스 조절
- 문제 설명
- 라이캣은 연말 출시할 게임의 게임 난이도를 조절하는 업무를 담당하게 되었습니다. 우선 능력치가 비정상적으로 높은 캐릭터들의 리스트를 뽑고자 합니다. 캐릭터 이름, 공격력, 방어력, 체력, 마력이 리스트로 ['Licat', 98, 30, 21, 60]와 같이 주어졌을 때 모든 능력치의 합이 350 이상이 되는 캐릭터의 이름을 출력하는 solution함수를 완성해주세요.

제한 사항
- 3 ≤ 캐릭터의 수 ≤ 10
- 10 ≤ 능력치 ≤ 100
- 350이 넘는 캐릭터의 이름은 알파벳순으로 출력합니다.

image.png

```python
# 문제 3번 정답
def solution(data):
    result = []

    for character in data:
        name, attack, defense, health, mana = character
        total_stats = attack + defense + health + mana

        if total_stats >= 350:
            result.append(name)

    result.sort()  # 알파벳 순으로 정렬
    return result
```

문제 4.  [주간 최고 온도 기록 분석](https://100.pyalgo.co.kr/?page=18)
- 주간 최고 온도 기록 분석
- 문제 설명
- 일주일 동안의 최고 기온이 기록된 데이터가 주어집니다. 이 데이터에서 가장 높은 온도를 기록한 상위 3일을 찾아 'YY-MM-DD: 온도' 형식으로 반환하는 코드를 작성해주세요. 데이터는 'YYYY-MM-DD' 형식의 날짜와 해당 날짜의 최고 온도(섭씨)로 구성됩니다.

제한 사항
- 각 날짜의 최고 온도는 정수로 주어집니다.
- 기온 데이터는 최소 7개(일주일)에서 최대 30개(한달)까지 주어질 수 있습니다.
- 날짜는 중복되지 않습니다.
- 중복된 온도일 경우 날짜가 빠른 것이 앞에 정렬되어야 합니다.

image.png

```python
# 문제 4번 정답
def solution(data):
    sorted_data = sorted(data.items(), key=lambda x: (-x[1], x[0]))
    ranking3_days = sorted_data[:3]
    result = [f"{date[2:]}: {temperature}" for date, temperature in ranking3_days]

    return result
```
